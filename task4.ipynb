4.Display 10 Different Matplotlib Styles  in py

```python
import matplotlib.pyplot as plt
import numpy as np

# Create X and Y data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# List of 10 Matplotlib styles
styles = [
    'ggplot', 'seaborn-v0_8', 'fivethirtyeight', 'bmh', 'classic',
    'dark_background', 'Solarize_Light2', 'fast', 'grayscale', 'tableau-colorblind10'
]

# Loop through styles and plot
for style in styles:
    plt.style.use(style)
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, label='sin(x)')
    plt.title(f'Matplotlib Style: {style}')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
