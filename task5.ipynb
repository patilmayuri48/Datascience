1)Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.  
Then create a horizontal bar chart showing total runs scored in their debut 5 years.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = ['2003', '2007', '2011', '2015', '2019']
sachin_runs = [1200, 1000, 800, 0, 0]     # Assume retired before 2015
kohli_runs = [300, 800, 1100, 1300, 1000]
sehwag_runs = [900, 1100, 700, 0, 0]       # Retired before 2015
```

```python
x = np.arange(len(years))  # Label locations
width = 0.25  # Width of each bar

plt.figure(figsize=(10,6))
plt.bar(x - width, sachin_runs, width, label='Sachin')
plt.bar(x, kohli_runs, width, label='Kohli')
plt.bar(x + width, sehwag_runs, width, label='Sehwag')

plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs scored over 5 years')
plt.xticks(x, years)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Sample total runs in debut 5 years
players = ['Sachin', 'Kohli', 'Sehwag']
debut_5_year_runs = [4500, 5000, 4300]

plt.figure(figsize=(8,5))
plt.barh(players, debut_5_year_runs, color=['skyblue', 'lightgreen', 'salmon'])

plt.xlabel('Total Runs')
plt.title('Total Runs in Debut 5 Years')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same using a horizontal bar chart and observe which is easiertounderstand

```python
labels = ['Android', 'iOS', 'Others']
market_share = [72, 26, 2]
```

```python
plt.figure(figsize=(6,6))
plt.pie(market_share, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Mobile OS Market Share')
plt.axis('equal')  # Keep circle shape
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.barh(labels, market_share, color=['limegreen', 'deepskyblue', 'gray'])

plt.xlabel('Market Share (%)')
plt.title('Mobile OS Market Share (Bar View)')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
